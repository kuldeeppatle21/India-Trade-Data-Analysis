### INDIA's IMPORT - EXPORT - PROFIT/LOSS WITH 15 COUNTRIES

```python
group_import =import_data.groupby(['country','year']).agg({'value':'sum'})
group_export =export_data.groupby(['country','year']).agg({'value':'sum'})
# group_import
```

```python
temp_import = group_import.groupby(['country']).agg({'value':'sum'})
temp_export = group_export.groupby(['country']).agg({'value':'sum'}).loc[temp_import.index.values]
# temp_export
```

```python
data_1=group_import.groupby(['country']).agg({'value':'sum'}).sort_values(by='value').tail(15)
data_2=temp_export
data_3=data_2-data_1
```

```python
data_1.columns=['Import']
data_2.columns=['Export']
data_3.columns=['Profit/Loss']
```

```python
df=pd.DataFrame(index=data_1.index.values)
df['Import'] = data_1
df['Export'] = data_2
df['Loss / Profit'] = data_3
```

```python
i_count = 0
e_count = 0
for i in range(len(df)):
    i_count = i_count + df.Import[i]
    e_count = e_count + df.Export[i]
print(i_count)
print(e_count)
```

<b>India imports more in comparison of export</b>

### YEAR-WISE TOP-10 COUNTRIES IMPORTING
